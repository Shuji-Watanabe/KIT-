```python
def Display_matrix_format01(a , c , Type00 ):
    M1 = np.array(a)
    M2 = M1.shape

    if(c == 1):
        Form01 = ''
    else:
        Form01 = '%s'

    if( Type00 == 'p' ):
        Form01 = Form01 + '\\begin{pmatrix} '
    else:
        if( Type00 == 'd'):
            Form01 = Form01 + '\\begin{vmatrix} ' 
        else:
            print("エラー:Disp_matrix_pmat")
            sys.exit()

    for j in range(0,M2[0],1) :
        for i in range(0,M2[1],1) :
            Form01 = Form01 + str(a[j][i])
            if (i != M2[1]-1 ):
                Form01 = Form01 + ' & '
            else:
                Form01 = Form01 + ' \\\\ '

    if( Type00 == 'p' ):
        Form01 = Form01 + ' \\end{pmatrix}'
    if( Type00 == 'd'):
        Form01 = Form01 + ' \\end{vmatrix}'
    return Form01

def Display_matrix_format02(a , c , Type00 ):
    M1 = np.array(a)
    M2 = M1.shape

    if(c == 1):
        Form01 = ''
    else:
        Form01 = '%s'

    Form01 = Form01 + '['
    for j in range(0,M2[0],1) :
        Form01 = Form01 + '['
        for i in range(0,M2[1],1) :
            Form01 = Form01 + str(a[j][i])
            if (i != M2[1]-1 ):
                Form01 = Form01 + ' , '
            else:
                Form01 = Form01 + ' ] '
        if (j != M2[0]-1 ):
            Form01 = Form01 + ' , '
        else:
            Form01 = Form01 + ' ] '
    return Form01

def Elementary_transformations(a,m,n,k0,Type):
    M1 = np.array(a)
    M2 = M1.shape
    if( Type == 1 ):
        a2 = M1
        for j in range(0,M2[0],1):
            for i in range(0,M2[1],1) :
                if(j==n-1):
                    a2[j][i] = a[j][i] + k0*a[m-1][i]
                else:
                    a2[j][i] = a[j][i]

        return a2,1

    if( Type == 2 ):
        a2 = M1
        for j in range(0,M2[0],1):
            for i in range(0,M2[1],1):
                if(j==m-1):
                    a2[j][i] = a[n-1][i]
                else:
                    if(j==n-1):
                        a2[j][i] = a[m-1][i] 
                    else:
                        a2[j][i] = a[j][i] 

        kk = -1
        return a2,kk

    if( Type == 3 ):
        a2 = M1
        for j in range(0,M2[0],1):
            for i in range(0,M2[1],1):
                if(j==m-1):
                    a2[j][i] = k0*a[m-1][i]
                else:
                    a2[j][i] = a[j][i] 
        kk = Rational(1,k0)

        return a2,kk

```
