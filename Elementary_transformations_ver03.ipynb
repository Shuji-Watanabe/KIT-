```python
def Display_matrix_format01(a,c,Type00):
    M1 = np.array(a)
    M2 = M1.shape
    #print("Check M1",M1,"Check M2",M2)
    if(c == 1):
        Form01 = ''
    else:
        Form01 = '%s'

    if( Type00 == 'p' ):
        Form01 = Form01 + '\\begin{pmatrix} '
    else:
        if( Type00 == 'd'):
            Form01 = Form01 + '\\begin{vmatrix} ' 
        else:
            print("エラー:Disp_matrix_pmat")
            #sys.exit()

    for j in range(0,M2[0],1) :
        for i in range(0,M2[1],1) :
            Form01 = Form01 + str(a[j][i])
            if (i != M2[1]-1 ):
                Form01 = Form01 + ' & '
            else:
                Form01 = Form01 + ' \\\\ '

    if( Type00 == 'p' ):
        Form01 = Form01 + ' \\end{pmatrix}'
    if( Type00 == 'd'):
        Form01 = Form01 + ' \\end{vmatrix}'
    return Form01

def Display_matrix_format02(a , c , Type00 ):
    M1 = np.array(a)
    M2 = M1.shape

    if(c == 1):
        Form01 = ''
    else:
        Form01 = '%s'

    Form01 = Form01 + '['
    for j in range(0,M2[0],1) :
        Form01 = Form01 + '['
        for i in range(0,M2[1],1) :
            Form01 = Form01 + str(a[j][i])
            if (i != M2[1]-1 ):
                Form01 = Form01 + ','
            else:
                Form01 = Form01 + ']'
        if (j != M2[0]-1 ):
            Form01 = Form01 + ','
        else:
            Form01 = Form01 + ']'
    return Form01
```

```python
def Elementary_transformations(a,m,n,k,Type_cal):
    M1 = np.array(a)
    M2 = M1.shape
    type_0 = Type_cal
    if( Type_cal >= 4 ):
        Type_cal = Type_cal - 3
        M1 = np.transpose(M1)
        M2 = M1.shape   
        
        a2 = Elementary_transformations_0(M1,m,n,k,Type_cal)[0]
        kk = Elementary_transformations_0(M1,m,n,k,Type_cal)[1]
        if( type_0 >= 4 ):
            a2 = np.transpose(a2)
        
    else:
        a2 = Elementary_transformations_0(M1,m,n,k,Type_cal)[0]
        kk = Elementary_transformations_0(M1,m,n,k,Type_cal)[1]
        

    return a2,kk
    
def Elementary_transformations_0(a,m,n,k,Type_cal):
    M1 = np.array(a)
    M2 = M1.shape
    if( Type_cal == 1 ):
        a2 = M1
        for j in range(0,M2[0],1):
            for i in range(0,M2[1],1) :
                if(j==n-1):
                    a2[j][i] = factor( a[j][i] + k*a[m-1][i] )
                else:
                    a2[j][i] = a[j][i]
        kk= 1

    elif( Type_cal == 2 ):
        a2 = M1
        for j in range(0,M2[0],1):
            for i in range(0,M2[1],1):
                if(j==m-1):
                    a2[j][i] = a[n-1][i]
                else:
                    if(j==n-1):
                        a2[j][i] = a[m-1][i] 
                    else:
                        a2[j][i] = a[j][i] 

        kk = -1

    elif( Type_cal == 3 ):
        a2 = M1
        for j in range(0,M2[0],1):
            for i in range(0,M2[1],1):
                if(j==n-1):
                    a2[j][i] = factor( k*a[n-1][i] )
                else:
                    a2[j][i] = a[j][i] 
        kk = 1/k
        
    return a2,kk
```

```python
def F_Test02(a,c0,Type_disp,Type_Cal,m,n,k):
    try:
        x,y,z = symbols("x y z")
        a= sympify(a);c0 = sympify(c0)
        m= sympify(m); n = sympify(n);k= sympify(k)
        display( Math( "変形前" ))
        
        if(c0 == 1 ):
            display( Math( latex( Matrix(a) )))
        else:
            display( Math(  r'%s \cdot %s '% ( latex(c0) , latex( Matrix(a)))))
    except:
        display( Math( "各項目に適切な値，または行列を入力してください．" )) 
    else:
        ans ={'X行+k*Y行':1,'X行とY行の交換':2,'k*X行':3,'X列+k*Y列':4,'X列とY列の交換':5,'k*X列':6}
        Cal_out = Elementary_transformations(a,m,n,k,ans[Type_Cal])[0]
        c0 = c0*Elementary_transformations(a,m,n,k,ans[Type_Cal])[1]
        display( Math( "変形後" ))
        if(c0 == 1 ):
            display( Math( latex( Matrix(Cal_out) )))
        else:
            display( Math(  r'\left\{%s \right\} \cdot %s '% ( latex(c0) , latex( Matrix(Cal_out)))))
        display( Math( "得られた係数(再計算用)" ))
        print(sympify(c0))
        display( Math( "得られた行列(再計算用)" ))
        print(sympify(Cal_out))
```
